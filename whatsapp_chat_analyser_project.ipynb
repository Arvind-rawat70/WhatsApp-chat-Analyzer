```python
import re
import pandas as pd

# Open file and read data
with open('/content/drive/MyDrive/WhatsApp Chat with B.Tech Four 2023-27.txt', encoding='utf-8') as f:
    data = f.read()

# Updated regex pattern including AM/PM
pattern = r'(\d{1,2})/(\d{1,2})/(\d{2}), (\d{1,2}):(\d{2})\s?(AM|PM)'

# Extract date components as tuples
dates = re.findall(pattern, data)

# Convert each tuple into a formatted date string
dates_str = [
    f"{m.zfill(2)}/{d.zfill(2)}/{y.zfill(2)}, {h.zfill(2)}:{mi.zfill(2)} {ampm}"
    for m, d, y, h, mi, ampm in dates
]

# Extract messages using re.split.
# Note: re.split with capturing groups interleaves the groups with message text.
# The extraction method may need adjusting based on your file's exact format.
parts = re.split(pattern, data)
# Assuming your chat format is consistent, one way to get messages is to take every 7th element,
# since the regex has 6 groups. (This may need fine-tuning for your file.)
messages = parts[7::7]

# Ensure both lists are of the same length
min_len = min(len(messages), len(dates_str))
messages = messages[:min_len]
dates_str = dates_str[:min_len]

# Create DataFrame and convert strings to datetime
df = pd.DataFrame({'user_message': messages, 'message_date': dates_str})
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %I:%M %p')
df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
## separate users and message
users = []
messages_list = []
for message in df['user_message']:
  entry = re.split('([\w\W]+?):\s', message)
  if entry[1:]: ## user name
     users.append(entry[1])
     messages_list.append(entry[2])
  else:
    users.append('group_notifaication')
    messages_list.append(entry[0])
df['user'] = users
df['message'] = messages_list
df.drop(columns=['user_message'], inplace = True)

df.head()
```

```python
df['year'] = df['date'].dt.year
df.head()
```

```python
df['months'] = df['date'].dt.month_name()
```

```python
df['day'] = df['date'].dt.day
```

```python
df['hours'] = df['date'].dt.hour
df['mintues'] = df['date'].dt.minute
```
